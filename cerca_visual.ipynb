```python
import keras
import os
import cv2 as cv2
from params import get_params
from sklearn import preprocessing
import numpy as np

import matplotlib.pyplot as plt
from build_database import build_database

from keras.models import Model
from keras.preprocessing import image
from keras.applications.vgg19 import VGG19
from keras.applications.vgg19 import preprocess_input
```

```python
base_model = VGG19(weights='imagenet')
```

```python
model = Model(inputs=base_model.input, outputs=base_model.get_layer('fc2').output)
```

```python
params = get_params()
```

```python
readfile = os.path.join(params['root'],params['root_save'], params['image_lists'],params['split'] + '.txt')
```

```python
with open(readfile,'r') as f:
        image_list = f.readlines()
```

```python
list = [] #inicializa lista
for image_name in image_list:

        # Read image
        list.append(os.path.join(params['root'],params['database'],
                                     params['split'],
                                     'images',image_name.rstrip()))
       
```

```python
j=0
img = []
x = []
for j in range(len(image_list)):   
    img.append(image.load_img(list[j], target_size=(224, 224)))
    x.append(image.img_to_array(img[j]))
    x[j] = np.expand_dims(x[j], axis=0)
    x[j] = preprocess_input(x[j])
```

```python
features = []
s = 0
for s in range(len(image_list)):
    features = model.predict(x[s])
    print features.shape
```
